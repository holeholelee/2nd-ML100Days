# 檢視與處理 Outliers
### 為何會有 outliers, 常見的 outlier 原因
* 未知值，隨意填補 (約定俗成的代入)，如年齡常見 0,999
* 可能的錯誤紀錄/手誤/系統性錯誤，如某本書在某筆訂單的銷售量 = 1000 本

```python
# Import 需要的套件
import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

%matplotlib inline

# 設定 data_path
dir_data = './data'
```

```python
f_app = os.path.join(dir_data, 'application_train.csv')
print('Path of read in data: %s' % (f_app))
app_train = pd.read_csv(f_app)
app_train.head()
```

```python
# 先篩選數值型的欄位
"""
YOUR CODE HERE, fill correct data types (for example str, float, int, ...)
"""
dtype_select = [np.dtype('float64')]
#首先先選擇欄位為數字的(flaot64)，把它變成list之後再把它變成存成一個numeric_columns。
numeric_columns = list(app_train.columns[list(app_train.dtypes.isin(dtype_select))])
numeric_columns
print(len(numeric_columns))
```

```python
# 再把只有 2 值 (通常是 0,1) 的欄位去掉
tem=[]
for i in range(0,len(numeric_columns)):
    if len(app_train[numeric_columns[i]].unique()) != 2:
        tem.append(numeric_columns[i])
len(tem)
```

```python
value_range = [[] , [] , []]
for i in range(0 , len(numeric_columns)):
    value_range[0].append(numeric_columns[i])
    value_range[1].append(app_train[numeric_columns[i]].mean())
    value_range[2].append(app_train[numeric_columns[i]].max())
#轉置一下    
value_range = pd.DataFrame(value_range).T
value_range
```

```python
#幫每一column取個名字
value_range.columns = ['index','mean','max']
value_range
```

```python
# 最大值離平均與中位數很遠
print(app_train['AMT_INCOME_TOTAL'].describe())
#根據計算每個columns的數值顯示出來，利用該欄的指標進行排序，在累積加總
def CDF(series):
    return series.value_counts().sort_index().cumsum()
    
cdf = CDF(app_train['AMT_INCOME_TOTAL'])
```

### Hints: Emprical Cumulative Density Plot, [ECDF](https://zh.wikipedia.org/wiki/%E7%BB%8F%E9%AA%8C%E5%88%86%E5%B8%83%E5%87%BD%E6%95%B0), [ECDF with Python](https://stackoverflow.com/questions/14006520/ecdf-in-python-without-step-function)

```python

# 繪製 Empirical Cumulative Density Plot (ECDF)
"""
YOUR CODE HERE
"""


plt.plot(list(cdf.index), cdf/cdf.max())
plt.xlabel('Value')
plt.ylabel('ECDF')
plt.xlim([cdf.index.min(), cdf.index.max() * 1.05]) # 限制顯示圖片的範圍
plt.ylim([-0.05,1.05]) # 限制顯示圖片的範圍

plt.show()

# 改變 y 軸的 Scale, 讓我們可以正常檢視 ECDF
plt.plot(np.log(list(cdf.index)), cdf/cdf.max())
plt.xlabel('Value (log-scale)')
plt.ylabel('ECDF')

plt.ylim([-0.05,1.05]) # 限制顯示圖片的範圍

plt.show()
```
